## Encoding a simplicial complex using GUDHI

In this notebook, you will learn how to create a simplicial complex using the modules **gudhi** and visualize it  using the modules  **networkx.**  

```python
# We import all the libraries we need
import gudhi as gd
import networkx as nx
```

```python
# We define a simplicial complex via SimplexTree
st = gd.SimplexTree()
```

```python
# We add vertices
st.insert([0])
st.insert([1])
st.insert([2])
st.insert([3])
st.insert([4])
```

```python
# We add edges
st.insert([0,1])
st.insert([1,2])
st.insert([0,4])
st.insert([4,2])
st.insert([2,3])
st.insert([3,4])
```

```python
# We add triangle
st.insert([2,3,4])
```

# To visualize our simplicial complex we use networkx 

```python
# Function to draw
def VisualizeSimplicialComplex(simplicialcomplex, pos=None):
    '''
    Draw the 1-skeleton of the simplicial complex 'simplicialcomplex', in the plane.
    If 'pos==None', the position of the vertices are chosen at random, with to the 'spring_layout' of networkx. Thus 
    the funtion display the simplicial complex.
    If pos is difference from None the function do not display anything.   
    '''
    nodes=[]; edges=[]
    for filtr in simpcomplex.get_skeleton(1):
        simplex = filtr[0]
        if len(simplex)==1:
            nodes.append(simplex[0])
        if len(simplex)==2:
            edges.append(simplex)
    G = nx.Graph()
    G.add_nodes_from(nodes)
    G.add_edges_from(edges)
    if pos == None:
        pos = nx.spring_layout(G)
        nx.draw(G, pos=pos, with_labels=True, font_weight='bold', node_color='magenta')
```

```python
VisualizeSimplicialComplex(st)
```
